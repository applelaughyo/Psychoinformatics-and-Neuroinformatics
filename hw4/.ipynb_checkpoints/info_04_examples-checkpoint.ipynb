#### 2.1.1 Example 1:  
<p><b>1</b>2<b>3</b></p>

```python
data='<p><b>1</b>2<b>3</b></p>'
```

##### 2.1.1.1 Using LXML

```python
import lxml.html
t=lxml.html.fromstring(data)
print(t,t[0],t[1])
print(t.text,t[0].text,t[1].text)
```

```python
print(t.xpath('//p'))
print(t.xpath('//p')[0].text_content()) #123
print(t.xpath('//p')[0].getchildren()) #1,3
print(t.xpath('//b')) #1,3
print(t.xpath('//p/text()')) #2
```

##### 2.1.1.2 Using Scrapy Selector

```python
from scrapy.selector import Selector
```

```python
t=Selector(text=data)
print(t)
```

```python
t.xpath('//b').extract()
```

```python
print(t.xpath('//p').extract())
print(t.xpath('//p/text()').extract())
print(t.xpath('//p/text()').extract_first())
print(t.xpath('//p/*').extract())
print(t.xpath('//p').re('\d'))
```

```python
print(t.xpath('//b').extract())
print(t.xpath('//b/text()').extract())
print(t.xpath('//b/text()').extract_first())
```

##### 2.1.3.2 Using Scrapy

```python
import scrapy
from scrapy.crawler import CrawlerProcess
```

```python
class MySpider(scrapy.Spider):
    name="ptt"
    start_urls=["https://www.google.com/search?q=the+difference+between+data+engineer+and+data+scientist"]
    def parse(self, response):
        for link in response.xpath('//a'):
            print(link.xpath('text()').extract())
            print(link.xpath('@href').extract())
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})
process.crawl(MySpider)
process.start()
```

#### 2.1.4 Analyzing texts from Google

##### 2.1.4.1 Using LXML

```python
import requests
import lxml.html
URI='https://www.google.com/search?q=the+difference+between+data+engineer+and+data+scientist'
r=requests.get(URI)
```

```python
print(r.text)
```

```python
t=lxml.html.fromstring(r.text)
x=t.xpath('//div[@class="BNeawe s3v9rd AP7Wnd"]/text()')
print('\n\n'.join(x))
```

```python
import numpy as np
x=t.xpath('//div[@class="BNeawe s3v9rd AP7Wnd"]')

for e in x:
    print(e.text_content(),'\n')
```

```python
x=t.xpath('//div[@class="BNeawe s3v9rd AP7Wnd"]')

for e in x:
    child=e.getchildren()
    content=e.text_content()
    #print(c,'\n')
    #print(content,'\n\n')
    if((len(child)==2) & (content[0:4]=="2022")):
        print(content)
        #print(e.getparent().getparent().getparent().getparent().getparent().getprevious().attrib.get('class'))
        print(e.getparent().getparent().getparent().getparent().getparent().getprevious().getchildren()[0].attrib.get('href'))
        print('\n')
```

##### 2.1.1.2 Using Scrapy

```python
import scrapy
from scrapy.crawler import CrawlerProcess
        
class MySpider(scrapy.Spider): #answers
    name="google"
    start_urls=["https://www.google.com/search?q=the+difference+between+java+and+javascript","https://www.google.com/search?q=the+difference+between+data+engineer+and+data+scientist"]
    def parse(self, response):
        print('\n'.join(response.xpath('//div[@class="BNeawe s3v9rd AP7Wnd"]/text()').extract()))
        
from scrapy.utils.project import get_project_settings
get_project_settings()
```

```python
process = CrawlerProcess(get_project_settings())
process.crawl(MySpider)
process.start()
```

### 3 Advanced Web Scrapping

#### 3.1.1 Sending HTML forms

##### 3.1.1.1 Using requests

```python
# Not working without sending required info:
import requests
URI="https://www.ptt.cc/bbs/Gossiping"
r=requests.get(URI)
print(r.text)
```

```python
# Working with required information:
import requests
URI="https://www.ptt.cc/ask/over18"
r=requests.post(URI,data={'yes':'yes','from':'/bbs/Gossiping/'})
print(r.text)
```

##### 3.1.1.2 Using Scrapy

```python
import scrapy
from scrapy.http import Request, FormRequest
from scrapy.crawler import CrawlerProcess

class Spider(scrapy.Spider):
    name="ptt"
    
    def start_requests(self):
        yield Request(url='https://www.ptt.cc/ask/over18',callback=self.login)
        
    def login(self, response):
        yield FormRequest.from_response(
            response,
            formdata={'yes':'yes','from':'/bbs/Gossiping/'},
            callback=self.parse
        )
        
    def parse(self,response):
        print(response.text)
        
        
process = CrawlerProcess()
process.crawl(Spider)
process.start()
```

```python
import scrapy
from scrapy.http import Request, FormRequest
from scrapy.crawler import CrawlerProcess

class Spider(scrapy.Spider):
    name="ptt"
    
    def start_requests(self):
        yield Request(url='https://www.ptt.cc/ask/over18',cookies={'Cookie':'over18=1','User-Agent':'Mozilla/5.0'},callback=self.login)
        
    def login(self, response):
        yield FormRequest.from_response(
            response,
            formdata={'yes':'yes','from':'/bbs/Gossiping/'},
            callback=self.parse
        )
        
    def parse(self,response):
        print(response.text)
        
        
process = CrawlerProcess()
process.crawl(Spider)
process.start()
```

### 3.3 Solving CAPTCHA by OCR

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By

URI='https://irs.thsrc.com.tw/IMINT/'
driver=webdriver.Edge(service=Service("./msedgedriver"))
driver.get(URI)
btn=driver.find_element(By.ID,"cookieAccpetBtn")
btn.click()
img = driver.find_element(By.XPATH,'//img[@id="BookingS1Form_homeCaptcha_passCode"]')
src=img.get_attribute('src')

r=requests.get(src,headers={'User-Agent':'Mozilla/5.0'})
with open('captcha.jpg', 'wb') as f:
    f.write(r.content)
    
from PIL import Image
img=Image.open('captcha.jpg')
img
```

```python
#!pip install easyocr
import easyocr
reader = easyocr.Reader(['en'])
```

```python
reader.readtext('captcha.jpg')
```

```python
img2=img.point(lambda p: 255 if p > 90 else 0)
img2.save('captcha2.jpg')
img2
```
