## 1 Validating the findings about power poses (8 points)
Do <a href="https://www.youtube.com/watch?v=Ks-_Mh1QhMc">power poses</a> really work?
Let's analyze the data (keller2017.csv) from:

Keller, V. N., Johnson, D. J., & Harder, J. A. (2017). <a href="http://hpc.psy.ntu.edu.tw/info/01/KellerEtal2017.pdf">
Meeting your inner super (wo)man: are power poses effective when taught?</a> Comprehensive Results in Social Psychology, 2(1), 106-122.

where<br>

4 experimental conditions were AC (Aware-Constrictive), AE (Aware-Expansive), UC (Unaware-Constrictive), & UE (Unaware-Expansive)<br>
<br>
and<br>
<br>
3 dependent measures were body expansiveness, performance (in a job interview), & hireability scored by 4 independent raters.

Please construct 3 pandas dataframes df1, and df2, and df3 where <br><br>
<b>df1</b>, derived from df, has only 13 columns of interest (cond, expansiveness1-4, performance1-4, hire1-4) and no missing values (2 points) ;<br><br>
<b>df2</b>, derived from df1, has 4 columns that average expansiveness1-4 into "expansiveness", "performance1-4" into "performance", & hire1-4 into "hire" (df1 & df2 should have the same number of rows) (2 points) ; <br><br>
<b>df3</b>, derived from df2, calculates the means of each dependent variable in each condition (2 points) ;

Finally, run three two-way ANOVAs to see if there is any statistically significant result for each dependent measure (2 point). See <a href="https://www.statology.org/two-way-anova-python/">this page</a> for an example. 

```python
import pandas as pd
df=pd.read_csv("https://ceiba.ntu.edu.tw/course/4671ea/content/keller2017.csv")
```

```python
df
```

```python
# Please write your codes here:
df1=df[['cond','hire_1','hire_2','hire_3','hire_4','expansiveness_1','expansiveness_2','expansiveness_3','expansiveness_4','performance_1','performance_2','performance_3','performance_4']].dropna()
df2 = pd.DataFrame(df1[['cond']])
df2['hire']=df1[['hire_1','hire_2','hire_3','hire_4']].mean(axis=1) 
df2['expansiveness'] = df1[['expansiveness_1','expansiveness_2','expansiveness_3','expansiveness_4']].mean(axis=1)
df2['performance'] = df1[['performance_1','performance_2','performance_3','performance_4']].mean(axis=1)

df3 = df2.groupby('cond').mean()
print(df1.shape,df2.shape)
print(df3)
```

## hire

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols


df2[['aware', 'expansive']] = df2['cond'].apply(lambda cond: pd.Series(list(cond)))

# 看 aware or not / expansive or not 對三個變數的影響

# hire
model_hire = ols('hire ~ aware + expansive + aware:expansive', data=df2).fit()
stats_hire = sm.stats.anova_lm(model_hire, typ=2)

print('Hire:')
print(stats_hire, '\n')
```

Since the p-values for aware and expansive are both > 0.05, this means that both factors don't have a statistically significant effect on hireability score.

And since the p-value for the interaction effect = 0.484093 > 0.05, this tells us that there is no significant interaction effect between aware and expansive.

## expansiveness

```python
#perform two-way ANOVA
# exp
model_exp = ols('expansiveness ~ aware + expansive + aware:expansive', data=df2).fit()
stats_exp = sm.stats.anova_lm(model_exp, typ=2)
print('Expansiveness:')
print(stats_exp, '\n')
```

Since the p-values for aware and expansive are both > 0.05, this means that both factors don't have a statistically significant effect on body expansiveness.

And since the p-value for the interaction effect = 0.445487 > 0.05, this tells us that there is no significant interaction effect between aware and expansive.

## performance

```python
# per
model_per = ols('performance ~ aware + expansive + aware:expansive', data=df2).fit()
stats_per = sm.stats.anova_lm(model_per, typ=2)
print('Performance:')
print(stats_per, '\n')
```

Since the p-values for aware and expansive are both > 0.05, this means that both factors don't have a statistically significant effect on performance.

And since the p-value for the interaction effect = 0.106852 > 0.05, this tells us that there is no significant interaction effect between aware and expansive.
